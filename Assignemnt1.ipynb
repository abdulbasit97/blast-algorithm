```python
def makeSequences(sequence,k):
    sequences = []
    for i in range(0,len(sequence)-k+1):
        sequences.append(sequence[i:i+k])
    return sequences
```

```python
makeSequences("aabbcc",4)
```

```python
import re
f = open("fasta.txt", "r")
textFile = f.read()
textFile = textFile.split('>')[1:]
```

```python
textFile
```

```python
sequences = []
proteins = []
for r in textFile:
    r = r.split('\n')
    proteins.append(r[0].split()[0]) 
    sequences.append("".join(r[1:]))
```

```python
sequences
```

```python
proteins
```

```python
k = int(input("Enter kmer: "))
kmersSeqeunces = {}
for i in range(0,len(sequences)):
    kmersSeqeunces[ proteins[i] ] = makeSequences(sequences[i],k)
```

```python
import json
with open('kmers.json', 'w') as f:
    json.dump(kmersSeqeunces, f)
```
